```python
import pandas as pd

df_debit = pd.read_csv('Customer_Segmentation.csv',usecols=['Customer Id','Age','Edu','Years Employed','Income','Card Debt','Other Debt','Defaulted','Address','DebtIncomeRatio'])
```

```python
import seaborn as sns

sns.scatterplot(data = df_debit, x = 'Age', y = 'Income', hue = 'DebtIncomeRatio')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df_debit[['Income', 'Age']], df_debit[['DebtIncomeRatio']], test_size=0.33, random_state=0)
```

```python
from sklearn import preprocessing

X_train_norm = preprocessing.normalize(X_train)
X_test_norm = preprocessing.normalize(X_test)
```

```python
from sklearn import KMeans

kmeans = KMeans(n_clusters = 3, random_state = 0, n_init='auto')
kmeans.fit(X_train_norm)
```
